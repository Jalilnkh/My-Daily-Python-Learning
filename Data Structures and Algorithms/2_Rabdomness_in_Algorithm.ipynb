# Randomization in Alorithm
Randomization in algorithms is a technique that uses randomness, like flipping a coin, to guide its logic and improve efficiency or simplify its design.
So whenever we want to have randomization inside our algorithm we can use this method and it can help us to have it. 


### Simple randomization in algorithm

```python
import random

def flip():
    gen_num = random.randint(1,6)
    return gen_num

def flip_toss(first_gen=None, counter=0):
    if first_gen is None:
        # First call - establish the target number
        first_gen = flip()
        counter = 1
    
    next_gen = flip()
    if next_gen == first_gen:
        # Same number, recurse with incremented counter
        return flip_toss(first_gen, counter + 1)
    else:
        # Different number, return result
        return counter, first_gen
```

```python
# how many time a toss number could happen sequantially in 1000 tosing 
def all_toss_shown_once():
    collect_info = {}
    for i in range(1000):
        counter, shown_toss = flip_toss()
        if counter > 1:
            collect_info[shown_toss] = collect_info.get(shown_toss, 0) + counter
    
    return collect_info           
```

```python
# Run the function and see results
result = all_toss_shown_once()
sorted_by_keys = dict(sorted(result.items()))

print(sorted_by_keys)

# Desired output something like this 
# {1: 60, 2: 48, 3: 52, 4: 68, 5: 48, 6: 61}
```

## Quicksort Algorithm

Quicksort is a method that breaks a problem into smaller parts. It picks one item called a "pivot" from the list and then splits the other items into two groups: those smaller than the pivot and those bigger than the pivot.

### Pseudo-code

        pivot = A[r]
        i = p-1
        for j = p to r-1 {
            if A(j) < pivot {
            i=i+1

        swap (A[i+1], A[r])

## Quickselect Algorithm and its Applications

Here we want to select or find k_th smallest element from a array.
To find these kind of values we can use several ideas:
- Naive one: Sort the array and then take the k_th element.
- Brute force: other one is brute force or using for loop by looking one by one into array.
- QuickSelect: use pivot and partition into 2 parts then select the k_th value.
    - In worst case we will get theta(n^2)
    - In random choice of pivot I will get theta(n) and it is really fast.
    




```python

data = [-7, 3, 5, -4, 6, 8, 12, 9, 15, 7]
k = 6
result = quickselect(data[:], k)
print("6th smallest =", result)
```
