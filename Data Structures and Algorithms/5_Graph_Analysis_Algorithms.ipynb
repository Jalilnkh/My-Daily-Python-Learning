# Breadth‑First Search (BFS)
Let's work on bfs, and implement its code.

```python
from typing import Dict, List, Hashable, Iterable
from script.bfs import bfs_order, bfs_shortest_path
Graph = Dict[Hashable, Iterable[Hashable]]
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}
print(bfs_order('A', graph))  # ['A', 'B', 'C', 'D', 'E', 'F']
```

```python

path, dist = bfs_shortest_path('A', 'F', graph)
print(path)  # ['A', 'C', 'F']
print(dist)  # distances from 'A' to all visited nodes
```
