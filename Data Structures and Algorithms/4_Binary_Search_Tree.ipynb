## Lets now solve a problem:

find k largest value in an unsorted array



```python
def find_k_longest(bst: BST, k: int) -> int:
    sorted_arr = bst.in_order_traversal()
    return sorted_arr[-k]

find_k_longest(bst, k=3)
```
