## Lets now solve a problem:

find k largest value in an unsorted array



```python
def find_k_longest(bst: BST, k: int) -> int:
    sorted_arr = bst.in_order_traversal()
    return sorted_arr[-k]

find_k_longest(bst, k=3)
```

## Red-Black Trees Basics
A Red-Black Tree is a self-balancing binary search tree that ensures time complexity for search, insertion, and deletion by coloring nodes red or black and enforcing specific rules.
We want to have a self balance tree. 
It is a BST plus extra stuff with it.

Rules in Red black trees:

- Node Color: Every node is colored either red or black WsCube Tech.
- Root Rule: The root node is always black Naukri.com.
- Red Rule: Red nodes cannot have red children (i.e., red nodes cannot be adjacent).
- Path Rule: Every path from a node (including the root) to any of its descendant NULL/NIL nodes must contain the same number of black nodes University of Wisconsin–Madison.
- Leaf Rule: All leaf nodes (NIL/NULL sentinel nodes) are considered black Scribd.
